# Procesamiento de imagenes por individuo #

```python
## Librerias necesarias
from scipy import misc
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
%matplotlib inline
```

```python
## Funciones
def normalizacion(x):
    a = []
    for pixel in np.nditer(x):
        b=float(pixel)/x.max()
        a.append(b)
  
    B = np.reshape(a, (len(x),len(x)))
    return B

def reajuste(x):
    for pixel in np.nditer(x, op_flags=['readwrite']):
        if  pixel >=0.85:
            pixel[...]=1
        else: pixel [...]= pixel
        if 0.8 < pixel < 0.85:
            pixel[...]=random.uniform(0.8,0.9)
        else:
            pixel[...]=pixel
        if pixel <0.75:
            pixel[...]=random.uniform(0.1,0.2)
        else:
            pixel[...]=pixel
    return x
```

```python
## Individuo1
indi1 = cv2.imread('//home/user/MEGAsync/proyectos/AutoDanio/images/indi1/raw.JPG')
indi1 = cv2.cvtColor( indi1, cv2.COLOR_RGB2GRAY )
indi1 = indi1[270:430, 700:1500]
indi1 = ndimage.rotate(indi1, -6.5)
indi1 = indi1[50:150,250:350]
indi1 = normalizacion(indi1)
indi1 = reajuste(indi1)
plt.imshow(indi1, cmap=plt.cm.copper)
indi1.shape
```

```python
## Individuo 1 aleatorio
indi1_ale = np.array(indi1).reshape(-1,).tolist()
indi1_ale = np.array(indi1_ale)
indi1_ale = random.sample(indi1_ale, len(indi1_ale))
indi1_ale = np.reshape(indi1_ale, (100,100))
plt.imshow(indi1_ale, cmap=plt.cm.copper)
```

```python
## Individuo2
indi2 = cv2.imread('//home/user/MEGAsync/proyectos/AutoDanio/images/indi2/raw.JPG')
indi2 = cv2.cvtColor( indi2, cv2.COLOR_RGB2GRAY )
indi2 = indi2[2700:3000, 2500:2900]
indi2 = ndimage.rotate(indi2, +20.5)
indi2 = indi2[150:250,280:380]
indi2 = normalizacion(indi2)
indi2 = reajuste(indi2)
plt.imshow(indi2, cmap=plt.cm.copper)
indi2.shape
```
