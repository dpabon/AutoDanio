```python
## Parametros iniciales propuestos por Nakamasu et al.2009
cmsp = 2      # Tasa de producción de Ms
cmsd = 0.2    # Tasa de decaimiento de Ms
dms = 0.015   # Tasa de difusion de Ms
cxsp = 2      # Tasa de producción de Xs
cxsd = 0.2    # Tasa de decaimiento de Xs
dxs = 0.015   # Tasa de difusion de Xs
cxlp = 2.8    # Tasa de producción de Xl
cxld = 0.2    # Tasa de decaimiento de Xl
dxl = 0.3     # Tasa de difusion de Xl

# . c1 actua sobre el melanoforo. c2 actua sobre el xantoforo.  c2 es 10 veces mas grande que c1(relacion según nakamasu et al 2008)

c1 = 1.5
c2= 15
c3= 11
c4 = 1.5

cmdths = 0.5 #este valor determina la sensibilidad del melanoforo a la  accion de rango corto (en este caso alta, dado que el rango de valores del de mshort va de 0.4 a0.5)
cxdths =  1.5 # este valor determina la sensiblidad del xantoforo al efecto del rango corto de xshort, en la aparicion del mismo.                                                                                                       
cmdthl = 1.42
cmapps = 
cmappxl = 
cxapps = 1
cmappl = 1.4
cmapps = 0.4
cmdthl = 5  # sensibilidad del melanoforo a el rango de accion largo (aumento )

pmapp = 0.005
pxapp = 0.2
pmdth = 0.8 # probabilidad de muerte de los melanoforos, la probabilidad está asignada por la propocion del mel y empty en la lista random_mel_death
pxdth = 0.8

```

```python
#recordar que el pixel vacio = 1
# en este loop se cambia los pixeles con valor de uno a cero, para que cuando compute el valor de los vecinos los pixeles vacios no sumen 

def replace_pixel_empty(image_final):
    image_final.astype(int)
    shape = len(image_final), len(image_final[0])
    N  = [[0,]*(shape[0])  for i in range(shape[1])]
    image_final_for_neigbours = [[0,]*(shape[0])  for i in range(shape[1])]
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            image_final_for_neigbours [r][s]=image_final[r][s]
            if image_final_for_neigbours[r][s]==1:
                image_final_for_neigbours[r][s]=0
            else:
                image_final_for_neigbours[r][s]=image_final_for_neigbours[r][s]
            

    return image_final_for_neigbours
            
                 
    return N

image_final_for_neigbours = replace_pixel_empty(image_final)
```

```python
def compute_neigbours(image_final):
    image_final.astype(int)
    shape = len(image_final), len(image_final[0])
    N  = [[0,]*(shape[0])  for i in range(shape[1])]
    image_final_for_neigbours = replace_pixel_empty(image_final)
    for x in range(1,shape[0]-1):
        for y in range(1,shape[1]-1):
            N[x][y] = image_final_for_neigbours[x][y-1] \
                    + image_final_for_neigbours[x-1][y]            +image_final_for_neigbours[x+1][y]   \
                    + image_final_for_neigbours[x][y+1]
            
                 
    return N
```

```python
def delta_ms (image_final):
    N=compute_neigbours(image_final)
    shape = len(image_final), len(image_final[0])
    ms  = [[0,]*(shape[0])  for i in range(shape[1])]
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if image_final[r][s]<0.2:
                ms[r][s] = (((cmsp * image_final[r][s]) - (cmsd * image_final[r][s])) \
                        + (dms * (N[r][s] - (4* image_final[r][s]))))
            else:
                ms[r][s]= 0
            
    return ms
ms = delta_ms(image_final)
ms = np.array(ms, dtype=float)
ms.shape
print ms
#for i in ms:
  #  print i
```

```python
def delta_xs (image_final):
    N=compute_neigbours(image_final)
    shape = len(image_final), len(image_final[0])
    xs  = [[0,]*(shape[0])  for i in range(shape[1])]    
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if 0.8 <= image_final[r][s] < 0.9:
                xs[r][s] = (((cxsp * image_final[r][s]) - (cxsd * image_final[r][s])) \
                        + (dxs * (N[r][s] - (4* image_final[r][s]))))
            else:
                xs[r][s]= 0
    return xs
xs = delta_xs (image_final)

xs = np.array(xs, dtype=float)
#xs.shape
#print xs

#for i in xs:
  #  print i
```

```python
def delta_xl (image_final):
    N=compute_neigbours(image_final)
    shape = len(image_final), len(image_final[0])
    xl  = [[0,]*(shape[0])  for i in range(shape[1])]
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if 0.8 <=image_final[r][s] <=0.9:
                xl[r][s] = (((cxlp * image_final[r][s]) - (cxld * image_final[r][s])) \
                        + (dxl * (N[r][s] - (4* image_final[r][s]))))
            else:
                xl[r][s]=xl[r][s]
                
    return xl
xl  = delta_xl(image_final)
xl = np.array(xs, dtype=float)
#for i in xl:
   # print i
```

```python
def mshort (image_final):
    shape = len(image_final), len(image_final[0])
    ms = delta_ms (image_final)
    xs = delta_xs (image_final)
    mshort  = [[0,]*(shape[0])  for i in range(shape[1])]
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if image_final[r][s]<0.2:
                 mshort[r][s] = ((c1 * ms[r][s])+ (c2 * xs[r][s])) # esto da como resultado  un intervalo de accion de rango corto, estos son tenidos en cuenta  para  la supervivencia o muerte del melanoforo.
            else:
                mshort[r][s]=0
    return mshort

mshort = mshort (image_final)

mshort = np.array(mshort, dtype=float)
mshort.shape
print mshort

#for i in mshort:
  #  print i
        
```

```python

def xshort (image_final):
    xs= delta_xs (image_final)
    ms= delta_ms (image_final)
    shape = len(image_final), len(image_final[0])
    xshort  = [[0,]*(shape[0])  for i in range(shape[1])]
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if image_final [r][s]==0.75:
                xshort[r][s] = ((c3 * ms[r][s])+ (c4 * xs[r][s]))
            else:
                xshort[r][s]=image_final[r][s]
    return xshort
xshort = xshort (image_final)
xshort = np.array(xs, dtype=float)
print xshort
#for i in xshort:
  #  print i
    
# los valores mas altos de xshort determinan un mayor efecto de rango corto a causa del xantoforo
```
