```python
## Funciones
def normalizacion(x):
    a = []
    for pixel in np.nditer(x):
        b=float(pixel)/x.max()
        a.append(b)
  
    B = np.reshape(a, (len(x),len(x)))
    return B

def reajuste(x):
    for pixel in np.nditer(x, op_flags=['readwrite']):
        if  pixel >=0.85:
            pixel[...]=1
        else: pixel [...]= pixel
        if 0.8 < pixel < 0.85:
            pixel[...]=random.uniform(0.8,0.9)
        else:
            pixel[...]=pixel
        if pixel <0.75:
            pixel[...]=random.uniform(0.1,0.2)
        else:
            pixel[...]=pixel
    return x
```

```python
def compute_neigboursV(V):
    shape = len(U), len(U[0])
    NV  = [[0,]*(shape[0])  for i in range(shape[1])]
    for x in range(1,shape[0]-1):
        for y in range(1,shape[1]-1):
            NV[x][y] = V[x][y-1] \
                    + V[x-1][y]            +V[x+1][y]   \
                    + V[x][y+1]
            
    NV = np.array(NV,dtype=float)           
    return NV

#NV1 = compute_neigbours(V)
#print NV1
```

```python
def compute_neigboursU(U):
    shape = len(U), len(U[0])
    NU  = [[0,]*(shape[0])  for i in range(shape[1])]
    for x in range(1,shape[0]-1):
        for y in range(1,shape[1]-1):
            NU[x][y] = U[x][y-1] \
                    + U[x-1][y]            +U[x+1][y]   \
                    + U[x][y+1]
            
    NU = np.array(NU,dtype=float)           
    return NU

#NU1 = compute_neigbours(U)
#print NU1
```

```python
def delta_ms (U):
    NU1=compute_neigboursU(U)
    shape = len(U), len(U[0])
    ms  = [[0,]*(shape[0])  for i in range(shape[1])]
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
                ms[r][s] = (((cmsp * U[r][s]) - (cmsd * U[r][s])) \
                + (dms * (NU1[r][s] - (4* U[r][s]))))
    return ms
mss = delta_ms(U)
mss = np.array(mss, dtype=float)
#mss.shape
print mss
#for i in ms:
  #  print i
```

```python
def ratesV (V):
    shape = len(V), len(V[0])
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if (c4*V[r][s])+ c6<= 0:
                V[r][s]= 0
            elif 0<=c4*U[r][s]+ c6 <V[r][s]:
                V[r][s] = c4*U[r][s]+c6
            else:
                V[r]= V[r][s]
            return V
```

```python
# tasas realistas
def ratesU (U):
    shape = len(V), len(V[0])
    for r in range(1,shape[0]-1):
        for s in range(1,shape[1]-1):
            if U[r][s] <0:
                U[r][s]=0
            elif 0<=c1*V[r][s]+c3 <U[r][s]:
                U[r][s] = c1*V[r][s]+c3
            else:
                U[r][s]= U[r][s]
            return U
        
```

```python
#función ensamble 
a = 2.8e-4
b = 5e-3
tau = .1

k = -.005

size = 10 # size of the 2D grid
dx = 2./size  # space step
T =  100       
dt = .9 * dx**2/2  # time step
n = int(T/dt)

def laplacian(Z): #[laplaciano def (Z):]
    Ztop = Z[0:-2,1:-1] #[ZTOP = Z [0: -2,1 -1]]
    Zleft = Z[1:-1,0:-2] #[Zleft = Z [1: 1,0: -2]]
    Zbottom = Z[2:,1:-1] #[Zbottom = Z [2: 1: -1]]
    Zright = Z[1:-1,2:] #[Zright = Z [1: 1,2:]]
    Zcenter = Z[1:-1,1:-1]# [Zcenter = Z [1: 1,1: -1]]
    laplacian = (Ztop + Zleft + Zbottom + Zright - 4 * Zcenter) / dx**2

    return laplacian





t_start = time()

for i in range(T):
    
    sum_neighbours= compute_neigboursU(U)
    sum_neighboursV = compute_neigboursV(V)
    ms = delta_ms(U)
    xs = delta_xs(V)
    #xl = delta_xl(image_final)
    R_mshort = mshort(U) 
    R_short = xshort(V)
    P_deathmel = deathmel(U)
    P_appermel = appermel(P_deathmel)
    P_deathxan = deathxan(V)
    P_apperxan = apperxan(P_deathxan)
    U = P_appermel
    U = np.array(U, dtype=float)
    V = P_apperxan
    V = np.array(V, dtype=float)
    for i in range(n):
        # We compute the Laplacian of u and v.
        
        deltaU = laplacian(U)
        deltaV = laplacian(V)
        
        #We take the values of u and v inside the grid.
        Uc = U[1:-1,1:-1]
        Vc = V[1:-1,1:-1]
        
        # We update the variables.
        U[1:-1,1:-1], V[1:-1,1:-1] = \
        Uc + dt * (a * deltaU + Uc - Uc**3 - Vc + k), \
        Vc + dt * (b * deltaV + Uc - Vc) / tau
        
       # for Z in (U, V):
          #  Z[0,:] = Z[1,:]
           # Z[-1,:] = Z[-2,:]
        Uq= ratesU(U)
       # U = np.array(U, dtype=float)
        V= ratesV(V)
        #V = np.array(V, dtype=float)
           
                
           
print "time AutoDanio", time()-t_start     

```
